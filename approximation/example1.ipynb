```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score as det
```

```python
def f(x):
    return 0.3*x**2 - 0.2*x
```

```python
X = range(0, 1700, 34)
print(max(X))
```

```python
Y = []
for x in X:
    Y.append(f(x))
```

```python
noise = np.random.normal(np.average(Y), 0.1*max(Y), len(X))
```

```python
Y=[]
c = 0
for x in X:
    Y.append(f(x) + noise[c])
    c+=1
```

```python
plt.plot(X, Y)
plt.xlabel('ось X')
plt.ylabel('ось Y')
plt.show()
```

#### Аппроксимация

```python
from sklearn.linear_model import LinearRegression as lappr
```

```python
X1 = np.array(X).reshape((-1, 1))
Y1=np.array(Y)
```

```python
model = lappr()
model.fit(X1, Y1)
```

```python
plt.plot(X1, Y1)
plt.plot(X1, model.predict(X1))
plt.xlabel('ось X')
plt.ylabel('ось Y')
plt.show()
```

```python
print('Коэфициент детерминации', model.score(X1, Y1))
```
